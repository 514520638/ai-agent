```python
# 字典列表
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model="MiniMax-M3",
    api_key='sk-cp-GsTRU_GlFRX_WPY8cf8IqBTDtrVJP4zl4EWLZ8D7627lrDtV-MbWYnWWK7rmvT7ZRluuMtUL7VP2sCD2uq6SzkpIlY0EAJtCIHW2uSGAT4QKYxX9mJQmK8I',
    base_url="https://api.minimaxi.com/v1",
)

messages1 = [
    {"role": "system", "content": "你是一个友好的AI助手"},
    {"role": "user", "content": "你好，我叫小明"},
]
messages2 = [
    {"role": "user", "content": "我叫什么名字？"},
]

response1 = llm.invoke(messages1)
response2 = llm.invoke(messages2)
print(f"AI助手的回答1是：{response1.content}")
print(f"AI助手的回答2是：{response2.content}")

# 添加记忆
messages1.append({"role": "assistant", "content": response1.content})
messages1.append({"role": "user", "content": "我叫什么名字"})

response3 = llm.invoke(messages1)
print(f"AI助手的回答3是：{response3.content}")
```

```python
# 消息对象列表
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage

llm = ChatOpenAI(
    model="MiniMax-M3",
    api_key='sk-cp-GsTRU_GlFRX_WPY8cf8IqBTDtrVJP4zl4EWLZ8D7627lrDtV-MbWYnWWK7rmvT7ZRluuMtUL7VP2sCD2uq6SzkpIlY0EAJtCIHW2uSGAT4QKYxX9mJQmK8I',
    base_url="https://api.minimaxi.com/v1",
)

messages = [
    SystemMessage(content="你是一个数学家"),
    HumanMessage(content="解释一下微积分的概念")
]
# 第一次对话
response = llm.invoke(messages)
print(f"第一次对话:{response.content}")

# 添加记忆
messages.append(AIMessage(content=response.content))
messages.append(HumanMessage(content="我刚刚问了什么问题"))

# 第二次对话
response2 = llm.invoke(messages)
print(f"第二次对话:{response2.content}")
```
